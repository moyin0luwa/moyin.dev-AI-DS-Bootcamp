```python
print("=" * 50)
print("Variables in Python")
print("=" * 50)

# Creating variables
age = 20
name = "Alex"

# Displaying variables
print(f"Age: {age}")
print(f"Name: {name}")

# Checking data types
print(f"\nType of age: {type(age)}")
print(f"Type of name: {type(name)}")

print("\n‚úì Variables created successfully!")
```

```python
print("=" * 50)
print("Data Types in Python")
print("=" * 50)

# Integer (int) - whole numbers
age = 21
students_count = 100
temperature = -5

# Float - decimal numbers
height = 1.75
pi = 3.14159
price = 29.99

# String (str) - text
name = "Moyin"
course = "AI & Data Science"
greeting = 'Hello, World!'

# Boolean (bool) - True or False
is_student = True
has_completed = False
is_enrolled = True

# Display all examples
print("\n--- INTEGER Examples ---")
print(f"age = {age} (type: {type(age).__name__})")
print(f"students_count = {students_count} (type: {type(students_count).__name__})")
print(f"temperature = {temperature} (type: {type(temperature).__name__})")

print("\n--- FLOAT Examples ---")
print(f"height = {height} (type: {type(height).__name__})")
print(f"pi = {pi} (type: {type(pi).__name__})")
print(f"price = {price} (type: {type(price).__name__})")

print("\n--- STRING Examples ---")
print(f"name = '{name}' (type: {type(name).__name__})")
print(f"course = '{course}' (type: {type(course).__name__})")
print(f"greeting = '{greeting}' (type: {type(greeting).__name__})")

print("\n--- BOOLEAN Examples ---")
print(f"is_student = {is_student} (type: {type(is_student).__name__})")
print(f"has_completed = {has_completed} (type: {type(has_completed).__name__})")
print(f"is_enrolled = {is_enrolled} (type: {type(is_enrolled).__name__})")

print("\n‚úì Python is dynamically typed - no need to declare types!\n")
```

```python
print("=" * 50)
print("Conditionals (Decision Making)")
print("=" * 50)

# Basic if-else example
age = 18

print(f"\nChecking eligibility for age: {age}")
if age >= 18:
    print("‚úì You are eligible")
else:
    print("‚úó You are not eligible")

# Using elif for multiple conditions
print("\n--- Grade Classification Example ---")
score = 85

if score >= 90:
    print(f"Score: {score} ‚Üí Grade: A (Excellent!)")
elif score >= 80:
    print(f"Score: {score} ‚Üí Grade: B (Very Good!)")
elif score >= 70:
    print(f"Score: {score} ‚Üí Grade: C (Good)")
elif score >= 60:
    print(f"Score: {score} ‚Üí Grade: D (Pass)")
else:
    print(f"Score: {score} ‚Üí Grade: F (Fail)")

# Multiple conditions with logical operators
print("\n--- Multiple Conditions Example ---")
age = 22
is_student = True

if age >= 18 and is_student:
    print("‚úì You qualify for student discount!")
elif age >= 18 and not is_student:
    print("‚úì You are an adult but no student discount")
else:
    print("‚úó You must be 18 or older")

print("\n‚ö†Ô∏è Remember: Python uses INDENTATION to define code blocks!\n")
```
